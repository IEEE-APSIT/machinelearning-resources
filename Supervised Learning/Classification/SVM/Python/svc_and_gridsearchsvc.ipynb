```python
import pandas as pd 
import numpy as np 
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.datasets import load_breast_cancer 
from sklearn.svm import SVC 
```

```python
cancer = load_breast_cancer() 
  
#dataset dict form me hai 
print(cancer.keys()) 
```

```python
print(cancer['DESCR'])
```

```python

#isme sab input data lelenge
df = pd.DataFrame(cancer['data'], 
					columns = cancer['feature_names']) 

# cancer column chaiye main as output variable
df_target = pd.DataFrame(cancer['target'], columns =['Cancer']) 

print("Feature Variables: ") 
print(df.info()) 
```

```python
df_target  #output column
```

```python
df #input column
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(df, np.ravel(df_target), test_size = 0.33, random_state = 101) 
```

```python
# train karege normal pe
model = SVC() 
model.fit(X_train, y_train) 

# comparison
y_pred = model.predict(X_test) 
print(classification_report(y_test, y_pred)) 
```

```python
#ab parameters search karege grid search ke saath model man
from sklearn.model_selection import GridSearchCV 

# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001]} 

#question me kernel nai tha parameter i have added it on my own above
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 5) 

# fitting the model for grid search 
grid.fit(X_train, y_train) 
```

```python
# print best parameter after tuning 
print(grid.best_params_) 

# print how our model looks after hyper-parameter tuning 
print(grid.best_estimator_) 
```

```python
from sklearn.metrics import confusion_matrix
```
